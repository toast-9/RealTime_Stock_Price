# Real Time Stock Price Analysis


```python
import yfinance as yf
import plotly.graph_objs as go

# Get the stock symbol from the user
stock_symbol = input("Enter the stock symbol (e.g., AAPL): ")

# Interval required 5 minutes
data = yf.download(tickers=stock_symbol, period='1d', interval='5m')

# Declare figure
fig = go.Figure()

# Candlestick
fig.add_trace(go.Candlestick(x=data.index,
                             open=data['Open'],
                             high=data['High'],
                             low=data['Low'],
                             close=data['Close'],
                             name='market data'))

# Add title
fig.update_layout(title=f'{stock_symbol} Live Share Price Evolution', yaxis_title='Stock Price (USD per share)')

# X-axes
fig.update_xaxes(rangeslider_visible=True,
                rangeselector=dict(
                    buttons=list([
                        dict(count=15, label='15m', step='minute', stepmode='backward'),
                        dict(count=45, label='45m', step='minute', stepmode='backward'),
                        dict(count=1, label='HTD', step='hour', stepmode='todate'),
                        dict(count=3, label='3h', step='hour', stepmode='backward'),
                        dict(step='all')
                    ]))
                )

fig.show()
```
